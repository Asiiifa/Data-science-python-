```python
import numpy as np
data = [12, 15, 14, 10, 15, 18, 20, 15, 12, 17]
mean = np.mean(data)
median = np.median(data)

mode = max(data, key=data.count)
print(f"Mean: {mean}, Median: {median}, Mode: {mode}")
```

```python



variance = np.var(data)

print(f"Variance: {variance}")
```

```python
stddev = np.std(data)

print(f"Standard Deviation: {stddev}")
```

```python
mad = np.mean(np.abs(data - np.mean(data)))

print(f"Mean Absolute Deviation (MAD): {round (mad, 2)}")
```

```python
cv =(stddev / mean) * 100

print(f"Coefficient of Variation (CV): {cv:.2f}%")
```

```python
Q1 = np.percentile(data, 25)

Q3=np.percentile (data, 75)

IQR =Q3 -Q1

print(f"Inter-Quartile Range (IQR): {IQR}")
```

```python
z_scores = [(x - mean) / stddev for x in data]

print("Z-Scores:")

for score in zip(data, z_scores):

 print(score)
```

```python
X =[12, 15, 14, 10, 15, 18, 20, 15, 12, 17]

Y=[22, 25, 19, 14, 20, 24, 28, 21, 18, 23]



cov_matrix = np.cov (X, Y)





covariance = cov_matrix[0, 1]

print(f"Covariance: {covariance}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns

population =np.random.exponential(scale=2, size=10000)



sns.histplot(population)
```

```python

sample_size = 30
num_samples = 1000


sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]



plt.hist(sample_means, bins=30, edgecolor='k', alpha=0.7, density=True)
plt.ylabel('Frequency')
plt.xlabel('Sample Mean')
plt.title('Demonstration of Central Limit Theorem (CLT)')
plt.grid(True)
plt.show()
```

```python

population = np.random.normal(loc=50, scale=15, size = 100000)


true_mean = np.mean(population)


sample_sizes = np.arange(1, 5000, 50)
sample_means = [np.mean(np.random.choice(population, size)) for size in sample_sizes]


plt.figure(figsize=(8, 5))
plt.plot(sample_sizes, sample_means, label ="Sample Mean", color='b', alpha=0.7)
plt.axhline(true_mean, color='r', linestyle ="dashed", label="True Population Mean")
plt.xlabel("Sample Size")
plt.ylabel("Mean Value")
plt.title("Law of Large Numbers Demonstration")
plt.legend()
plt.grid(True)
plt.show ()


```

```python
from scipy.stats import bernoulli

p=0.5

data = bernoulli.rvs(p, size=1000)
plt.hist(data, bins=[-0.5, 0.5, 1.5], edgecolor='black', alpha=0.7, density=True)

plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])

plt.xlabel('Outcome')

plt.ylabel('Probability')

plt.title(f'Bernoulli Distribution (p={p})')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
from scipy.stats import binom

n= 10

p=0.5

data = binom.rvs(n, p, size=1000)



plt.hist(data, bins=np.arange(n+2)-0.5, edgecolor="black", alpha=0.7, density=True)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')

plt.title(f'Binomial Distribution (n={n}, p={p})')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.xticks(range(n+1))

plt.show()
```

```python
from scipy.stats import poisson
lambda_=5



data =poisson.rvs(mu=lambda_, size=1000)

plt.hist(data, bins=range(0, max(data)+1), edgecolor='black', alpha=0.7, density=True)

plt.xlabel('Number of Events')

plt.ylabel('Probability')
plt.title(f'Poisson Distribution (A={lambda_})')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(range(0, max(data)+1))

plt.show()
```

```python
from scipy.stats import norm


mu = 58
sigma = 10 # Standard Deviation

data = np.random.normal(mu, sigma, 1000) #

plt.hist(data, bins=30, edgecolor='black', alpha=0.7, density=True)




x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

plt.plot(x, norm.pdf(x, mu, sigma), 'r', label="Theoretical PDF")

plt.xlabel('Value')
plt.ylabel('Density')
plt.title(f'Normal Distribution (={mu}, ={sigma})')
plt.legend()

plt.grid(True)
plt.show()
```

```python
from scipy.stats import expon

lambda_ = 1.5
mean = 1/lambda_

data = np.random.exponential(scale=mean, size=1000)
plt.hist(data, bins=30, edgecolor="black", alpha=0.7, density=True)


x = np.linspace(0, max(data), 100)
plt.plot(x, expon.pdf(x, scale=mean), 'r', label="Theoretical PDF")
plt.xlabel('Time Between Events')
plt.ylabel("Density")
plt.title(f'Exponential Distribution (={lambda_})')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import uniform

a, b = 10, 20

data = np.random.uniform(a, b, 1000)

plt.hist(data, bins = 30, edgecolor='black', alpha=0.7, density=True)



x = np.linspace(a, b, 100)

plt.plot(x, uniform.pdf(x, loc=a, scale=b-a), 'r', label="Theoretical PDF")

plt.xlabel('Value')

plt.ylabel('Density')

plt.title(f'Continuous Uniform Distribution [{a}, {b}]')

plt.legend()

plt.grid(True)

plt.show()
```

```python
from scipy.stats import randint


a, b = 1, 6


data = np.random.randint(a, b + 1, 1000)


plt.hist(data, bins=np.arange(a, b + 2) - 0.5, edgecolor="black", alpha=0.7, density=True)


x = np.arange(a, b + 1)

plt.scatter(x, randint.pmf(x, a, b + 1), color='red', label="Theoretical PMF", zorder=3)

plt.xlabel('Value')

plt.ylabel('Probability')

plt.title(f'Discrete Uniform Distribution [{a}, {b}]')
plt.xticks(range(a, b + 1))

plt.legend()

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
