```python
#mergeing data in pandas
```

```python
a=pd.DataFrame({'id' :[1,2,3,4,5],
                'name' :['Alex',  'Allen','Alice','Ayoung','Alax'],
                'sub_id':[1, 2, 4, 6, 5]})
b = pd.DataFrame({'id': [1, 2, 3, 4, 5],
                  'Name': ["Betty", "Brian", 'Bryce', 'Billy', 'Bran'],
                  'Sub id': [2, 4, 3, 6, 5]})
```

```python
a
```

```python
b
```

```python
pd.merge(a,b,on='id') #how = inner (intersection)
```

```python
pd.merge(a,b,on = 'id')
```

```python
pd.merge(a, b, left_on='sub_id', right_on='Sub id')
```

```python
pd.merge(a,b,on= 'id', how = 'outer') # union
```
