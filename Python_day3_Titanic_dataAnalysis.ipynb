```python
#####   MERGING DATA IN PANDAS


a=pd.DataFrame({'id' :[1,2,3,4,5],
                'Name':['Alax', 'Amy', 'Allen', 'Alice', 'Ayoung'],
                'Sub_id': [1,2,3,4,5]})
b = pd.DataFrame({'id' :[1,2,3,4,5],
                'Name':['Betty', 'Brian', 'Bryce', 'Billy', 'Bran'],
                'Sub_id': [1,2,3,4,5]})

a

```

```python
pd.merge(a, b, on = 'id')          #### how = inner(intersection)
```

```python
pd.merge(a,b, on='Name')
```

```python
pd.merge(a,b, on='Sub_id')
```

```python
pd.merge(a,b, on='id', how = 'outer')     ###UNION
```

```python
pd.merge(a,b, on='Name', how = 'outer')
```

```python
pd.merge(a,b, on='Sub_id', how = 'outer')
```

```python
pd.merge(a,b, on='Sub_id', how = 'left')
```

```python
pd.merge(a,b, on='Sub_id', how = 'right')
```

```python
pd.merge(a,b, on='id', how = 'left')
```

```python

pd.merge(a,b, on=['id','Sub_id'], how = 'inner')
```

```python
pd.merge(a,b, on='Sub_id', how = 'inner')
```

```python
su = pd.merge(a,b, on='Sub_id', how = 'outer')
su
```
